
# Using Statsbomb
Getting familiar with Statsbomb data


```python
#importing SBopen class from mplsoccer to open the data
from mplsoccer import Sbopen
# The first thing we have to do is open the data. We use a parser SBopen available in mplsoccer.
parser = Sbopen()
```

## Competition data
Using method *competition* of the parser we can explore competitions to find the competition we are interested in.It gives information of the all the available competition.




```python
#opening data using competition method
df_competition = parser.competition()
#structure of data
#df_competition.info()
```

```python
df_competition.head()
```

we can see for the same competetion we have different seasons data

## Match data
Using method *match* of the parser we can explore matches of a competition to find the match we are interested in.
zooming in one level deeper to get all matches for that season of that competition.



```python
#opening data using match method
df_match = parser.match(competition_id=72, season_id=30)
#structure of data
#df_match.info()
```

```python
df_match.head()
```

```python
print(df_match['match_id'].nunique())
```

```python
print(df_match.shape)
```

there are totally 52 matches for this competition_id=72, season_id=30

## Lineup data
lineup is the details of all players involved in a specific match (this includes selected 11 + substitutes )
we can use lineup method and pass match id as the parameters, it returns a df that gives data of pplayers played for that match



```python
#opening data using match method
df_lineup = parser.lineup(22949)
#structure of data
df_lineup.head()

```

```python
players_per_country = df_lineup.groupby("country_name")["player_id"].nunique().reset_index()
players_per_country.columns = ["country_name", "num_players"]
players_per_country
```

from this we can understand, two countries (Germany and Spain) played against each other, with 22 and 23 players listed respectively (starters + substitutes).
