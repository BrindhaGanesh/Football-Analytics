## Competition data
 



```python
#path to data
compt_path = r'C:\footballanalysis\football\Footballdata\wyscout\competitions.json'

#open data 
with open(compt_path) as f:
    data = json.load(f)
    #save it in dataframe 
df_competitions = pd.DataFrame(data)
 
df_competitions.head()
```
